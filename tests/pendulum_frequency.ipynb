```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
loss_1 = torch.load('../loss/val_Pendulum_w0_1.pt')
loss_2 = torch.load('../loss/val_Pendulum_w0_2.pt')
loss_3 = torch.load('../loss/val_Pendulum_w0_3.pt')
loss_3_25 = torch.load('../loss/val_Pendulum_w0_3_25.pt')
loss_3_5 = torch.load('../loss/val_Pendulum_w0_3_5.pt')
loss_4 = torch.load('../loss/val_Pendulum_w0_4.pt')
```

```python
N = torch.arange(len(loss_1))
Epochs = torch.arange(1, len(loss_1) + 1)

plt.figure(figsize=(13, 7))
plt.plot(Epochs, loss_1, label='$\omega_0$ = 1', linestyle='-', color='blue', linewidth=0.5)
plt.plot(Epochs, loss_2, label='$\omega_0$ = 2', linestyle='-', color='red', linewidth=0.5)
plt.plot(Epochs, loss_3, label='$\omega_0$ = 3', linestyle='-', color='green', linewidth=0.5)
plt.plot(Epochs, loss_3_25, label='$\omega_0$ = 3.25', linestyle='-', color='brown', linewidth=0.5)
plt.plot(Epochs, loss_3_5, label='$\omega_0$ = 3.5', linestyle='-', color='purple', linewidth=0.5)
plt.plot(Epochs, loss_4, label='$\omega_0$ = 4', linestyle='-', color='orange', linewidth=0.5)
plt.yscale('log')

plt.legend(frameon=False, loc=(1.05, 0.75), fontsize=15, title='Frequency of Pendulum')
plt.xlabel('Epochs', fontsize=15)
plt.ylabel('Validation Loss', fontsize=15)
plt.title('Validation Loss for Pendulum with Different Frequencies', fontsize=15)
plt.grid()
plt.savefig("Loss_comparison.png", bbox_inches='tight', pad_inches=0.1, dpi=100, facecolor="white") 
plt.show()
```
