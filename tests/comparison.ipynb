```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def average_every_k(k, name):
    arr = np.load(name)
    n = (len(arr) // k) * k
    trimmed = arr[:n]
    reshaped = trimmed.reshape(-1, k)
    return reshaped.mean(axis=1)
```

```python
K = 500

N = len(MSE_1e0)
alpha = 0.7
width = 2

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(average_every_k(K, "data/mse_data_wode_5.0e+00.npy"), label = "$W_{ODE}$ = 5", alpha=alpha, lw=width)
ax.plot(average_every_k(K, "data/mse_data_wode_1.0e+00.npy"), label = "$W_{ODE}$ = 1", alpha=alpha, lw=width)
ax.plot(average_every_k(K, "data/mse_data_wode_5.0e-01.npy"), label = "$W_{ODE}$ = 0.5", alpha=alpha, lw=width)
ax.plot(average_every_k(K, "data/mse_data_wode_1.0e-01.npy"), label = "$W_{ODE}$ = 0.1", alpha=alpha, lw=width)
ax.plot(average_every_k(K, "data/mse_data_wode_1.0e-02.npy"), label = "$W_{ODE}$ = 0.01", alpha=alpha, lw=width)
ax.plot(average_every_k(K, "data/mse_data_wode_1.0e-03.npy"), label = "$W_{ODE}$ = 0.001", alpha=alpha, lw=width)
ax.plot(average_every_k(K, "data/mse_data_wode_0.0e+00.npy"), label = "$W_{ODE}$ = 0", alpha=alpha, lw=width)

ax.set_yscale('log')
ax.legend()
ax.grid()
ax.set_title("MSE for different $w_{ODE}$ values")
ax.set_xlabel(f"Averaged over every {K} epochs")
ax.set_ylabel("MSE")
plt.savefig("figures/MSE_ode_weight_comaprison.png", dpi=300, bbox_inches='tight')
plt.show()
```
