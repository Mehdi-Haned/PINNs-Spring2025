```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = 50

MSE_0e0 = np.load("data/mse_data_wode_0.0e+00.npy")[0:-1:M]
MSE_1e0 = np.load("data/mse_data_wode_1.0e+00.npy")[0:-1:M]
MSE_5e0 = np.load("data/mse_data_wode_5.0e+00.npy")[0:-1:M]
MSE_1e_1 = np.load("data/mse_data_wode_1.0e-01.npy")[0:-1:M]
MSE_5e_1 = np.load("data/mse_data_wode_5.0e-01.npy")[0:-1:M]
MSE_1e_2 = np.load("data/mse_data_wode_1.0e-02.npy")[0:-1:M]
MSE_1e_3 = np.load("data/mse_data_wode_1.0e-03.npy")[0:-1:M]


N = len(MSE_1e0)
alpha = 0.7
width = 2

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(range(N), MSE_5e0, label="$w_{ODE} = 5$", lw =width, alpha=alpha)
ax.plot(range(N), MSE_1e0, label="$w_{ODE} = 1$", lw =width, alpha=alpha)
ax.plot(range(N), MSE_5e_1, label="$w_{ODE} = 0.5$", lw =width, alpha=alpha)
ax.plot(range(N), MSE_1e_1, label="$w_{ODE} = 0.1$", lw =width, alpha=alpha)
ax.plot(range(N), MSE_1e_2, label="$w_{ODE} = 0.01$", lw =width, alpha=alpha)
ax.plot(range(N), MSE_1e_3, label="$w_{ODE} = 0.001$", lw =width, alpha=alpha)
ax.plot(range(N), MSE_0e0, label="$w_{ODE} = 0$", lw =width, alpha=alpha)

ax.set_yscale('log')
ax.legend()
ax.grid()
ax.set_title("MSE for different $w_{ODE}$ values")
ax.set_xlabel(f"every {M} epochs")
ax.set_ylabel("MSE")
plt.savefig("figures/MSE_ode_weight_comaprison.png", dpi=300, bbox_inches='tight')
plt.show()
```
