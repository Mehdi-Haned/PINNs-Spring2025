```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def average_every_k(k, name):
    arr = np.load(name)
    n = (len(arr) // k) * k
    trimmed = arr[:n]
    reshaped = trimmed.reshape(-1, k)
    return reshaped.mean(axis=1)
```

```python
k = 500

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(average_every_k(k, "data/error_SHO_ODE_VdP_DATA_0.001.npy"), label="$\mu$=0.001")
ax.plot(average_every_k(k, "data/error_SHO_ODE_VdP_DATA_0.002.npy"), label="$\mu$=0.002")
ax.plot(average_every_k(k, "data/error_SHO_ODE_VdP_DATA_0.005.npy"), label="$\mu$=0.005")
ax.plot(average_every_k(k, "data/error_SHO_ODE_VdP_DATA_0.010.npy"), label="$\mu$=0.01")
ax.plot(average_every_k(k, "data/error_SHO_ODE_VdP_DATA_0.020.npy"), label="$\mu$=0.02")
ax.plot(average_every_k(k, "data/error_SHO_ODE_VdP_DATA_0.050.npy"), label="$\mu$=0.05")

ax.set_title("Mean squared error of PINN and Van der Pol solution")
ax.set_xlabel(f"Every {k} steps")
ax.set_ylabel(f"Mean squared error (Averaged over {k} steps)")
ax.set_yscale('log')
ax.legend()
ax.grid()
ax.set_ylim(1e-6, 1e1)
plt.savefig("VdP_Data_SHO_ODE.png", dpi=300)
plt.show()
```
