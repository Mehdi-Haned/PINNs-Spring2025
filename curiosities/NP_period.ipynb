```python
import numpy as np
from scipy.special import ellipk
import matplotlib.pyplot as plt
```

```python
w0  = 2*np.pi*100
x0s = np.arange(0, np.pi, 1e-2)
ks  = np.sin(x0s/2)
Ts  = 4/w0 * ellipk(ks**2)

plt.plot(x0s, Ts, label="NP Period")
plt.plot(x0s, np.pi*2/w0 * np.ones_like(x0s), label="SHO Period")
plt.title("Period of nonlinear pendulum vs SHO \n $\omega_0 = 2\pi$")
plt.xlabel("Initial Position")
plt.ylabel("Period (s)")
plt.grid()
plt.legend()
plt.savefig("NP_period.png")
plt.show()
```
