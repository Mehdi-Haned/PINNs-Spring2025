```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def average_every_k(k, name):
    arr = np.load(name)
    n = (len(arr) // k) * k
    trimmed = arr[:n]
    reshaped = trimmed.reshape(-1, k)
    return reshaped.mean(axis=1)
```

```python
k = 100

mse_1    = average_every_k(k, "data/mse_SHO_ODE_NP_DATA_1.000.npy")
mse_0_75 = average_every_k(k, "data/mse_SHO_ODE_NP_DATA_0.750.npy")
mse_0_5  = average_every_k(k, "data/mse_SHO_ODE_NP_DATA_0.500.npy")
mse_0_25 = average_every_k(k, "data/mse_SHO_ODE_NP_DATA_0.250.npy")
mse_0_1  = average_every_k(k, "data/mse_SHO_ODE_NP_DATA_0.100.npy")
mse_0_05 = average_every_k(k, "data/mse_SHO_ODE_NP_DATA_0.050.npy")
mse_0_01 = average_every_k(k, "data/mse_SHO_ODE_NP_DATA_0.010.npy")
```

```python
lw = 2
alpha = 0.8

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(mse_1, label="$x_0 = 1.0$", linewidth=lw, alpha=alpha)
ax.plot(mse_0_75, label="$x_0 = 0.75$", linewidth=lw, alpha=alpha)
ax.plot(mse_0_5, label="$x_0 = 0.5$", linewidth=lw, alpha=alpha)
ax.plot(mse_0_25, label="$x_0 = 0.25$", linewidth=lw, alpha=alpha)  
ax.plot(mse_0_1, label="$x_0 = 0.1$", linewidth=lw, alpha=alpha)

ax.set_title("Mean squared error of PINN and Nonlinear Pendulum solution")
ax.set_xlabel(f"Every {k} steps")
ax.set_ylabel(f"Mean squared error (Averaged over {k} steps)")
ax.set_yscale('log')
ax.legend()
ax.grid()
plt.show()
```
