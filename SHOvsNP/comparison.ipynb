```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def average_every_k(k, name):
    arr = np.load(name)
    n = (len(arr) // k) * k
    trimmed = arr[:n]
    reshaped = trimmed.reshape(-1, k)
    return reshaped.mean(axis=1)
```

```python
k = 50

err_0_100 = average_every_k(k, "data/error_SHO_ODE_NP_DATA_0.100.npy")
err_0_150 = average_every_k(k, "data/error_SHO_ODE_NP_DATA_0.150.npy")
err_0_200 = average_every_k(k, "data/error_SHO_ODE_NP_DATA_0.200.npy")
err_0_250 = average_every_k(k, "data/error_SHO_ODE_NP_DATA_0.250.npy")
```

```python
lw = 2
alpha = 0.8

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(err_0_250, label="$x_0$=0.25")
ax.plot(err_0_200, label="$x_0$=0.2")
ax.plot(err_0_150, label="$x_0$=0.15")
ax.plot(err_0_100, label="$x_0$=0.1")

ax.set_title("Mean squared error of PINN and Nonlinear Pendulum solution")
ax.set_xlabel(f"Every {k} steps")
ax.set_ylabel(f"Mean squared error (Averaged over {k} steps)")
ax.set_yscale('log')
ax.legend()
ax.grid()
ax.set_ylim(1e-4, 1e0)
plt.show()
```
