```python
from PIL import Image

import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import matplotlib.ticker as ticker
torch.manual_seed(12)
```

```python
w01_E = torch.load('data/E_Training_info_w0=1.pt')
w01 = torch.load('data/Training_info_w0=1.pt')
w04_E = torch.load('data/E_Training_info_w0=4.pt')
w04 = torch.load('data/Training_info_w0=4.pt')
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
epochs = w01_E.size()[0]
ax.plot(range(epochs), w01_E[:, 6], label='x1 MSE energy $w_0 = 1$', color='blue')
ax.plot(range(epochs), w01[:, 6], label='x1 MSE no energy $w_0 = 1$', color='red')
ax.plot(range(epochs), w04_E[:, 6], label='x1 MSE energy $w_0 = 4$', color='blue', linestyle='--')
ax.plot(range(epochs), w04[:, 6], label='x1 MSE no energy $w_0 = 4$', color='red', linestyle='--')
ax.set_yscale('log')
ax.set_ylabel('MSE')
ax.set_title("MSEs for x1")
ax.set_ylim(1e-5, 1e1)
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
epochs = w01_E.size()[0]
ax.plot(range(epochs), w01_E[:, 7], label='x2 MSE energy $w_0 = 1$', color='blue')
ax.plot(range(epochs), w01[:, 7], label='x2 no MSE energy $w_0 = 1$', color='red')
ax.plot(range(epochs), w04_E[:, 7], label='x2 MSE energy $w_0 = 4$', color='blue', linestyle='--')
ax.plot(range(epochs), w04[:, 7], label='x2 MSE no energy $w_0 = 4$', color='red', linestyle='--')
ax.set_yscale('log')
ax.set_ylabel('MSE')
ax.set_title("MSEs for x2")
ax.set_ylim(1e-5, 1e1)
plt.legend()
plt.show()
```
