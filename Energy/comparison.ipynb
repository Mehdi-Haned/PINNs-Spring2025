```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
MSE_1e0 = np.load("data/MSE_history_1.0e+00.npy")
MSE_1e_1 = np.load("data/MSE_history_1.0e-01.npy")
MSE_1e_2 = np.load("data/MSE_history_1.0e-02.npy")
MSE_1e_3 = np.load("data/MSE_history_1.0e-03.npy")
MSE_5e_1 = np.load("data/MSE_history_5.0e-01.npy")
```

```python
N = len(MSE_1e0)
plt.figure(figsize=(10, 6))
plt.plot(np.arange(N), MSE_1e0, label="$w_{ODE}$ = 1e0")
plt.plot(np.arange(N), MSE_1e_1, label="$w_{ODE}$ = 1e-1")
plt.plot(np.arange(N), MSE_1e_2, label="$w_{ODE}$ = 1e-2")
plt.plot(np.arange(N), MSE_1e_3, label="$w_{ODE}$ = 1e-3")
plt.plot(np.arange(N), MSE_5e_1, label="$w_{ODE}$ = 5e-1")
plt.plot()
plt.grid()
plt.title("MSE Histories")
plt.legend()
plt.yscale("log")
plt.show()
```
